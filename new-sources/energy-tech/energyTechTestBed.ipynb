```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Setup Chrome options
options = Options()
options.add_argument('--headless')  # Run in headless mode
options.add_argument('--disable-gpu')

# Path to your ChromeDriver
driver_path = '/path/to/chromedriver'

# Start driver
service = Service(driver_path)
driver = webdriver.Chrome(service=service, options=options)

# Open the article URL
url = "https://www.energytech.com/energy-efficiency/article/21271715/comed-and-partners-awarded-45m-for-chicago-building-performance-hub"
driver.get(url)

# Let the page load
time.sleep(3)

# Extract title
title = driver.find_element(By.TAG_NAME, 'h1').text

# Extract publication date
pub_date = driver.find_element(By.CSS_SELECTOR, 'time').get_attribute('datetime')

# Extract article content
paragraphs = driver.find_elements(By.CSS_SELECTOR, 'article p')
article_text = "\n".join([p.text for p in paragraphs])

# Output the scraped data
print("Title:", title)
print("Published:", pub_date)
print("\nArticle Text:\n", article_text)

# Close the driver
driver.quit()
```
