```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse


def extract_article_text(full_article_url):
    alias_path = urlparse(full_article_url).path
    payload = {
        "query": "query getWebsiteLayoutPage($alias: String!, $useCache: Boolean, $preview: Boolean, $cacheKey: String) {\n  getWebsiteLayoutPage(\n    input: { alias: $alias, useCache: $useCache, preview: $preview, cacheKey: $cacheKey }\n  ) {\n    id\n    name\n    module\n    type\n    alias\n    contentTypes\n    pageType\n    isGlobal\n    tenants\n    propagate\n    hideHeader\n    hideFooter\n    key\n    loadMoreType {\n      type\n    }\n    primaryGrid\n    secondaryGrid\n    excludeAds {\n      welcomeAd\n      headerLeaderboardAd\n      stickyLeaderboardAd\n      contentBodyNativeAd\n      contentBodyEmbedAd\n      contentListNativeAd\n      reskinAd\n    }\n    pageData\n    cache\n    created\n    usedContentIds\n    usedIssueIds\n  }\n}",
        "variables": {
            "alias": alias_path,
            "cacheKey": "v2.8",
            "preview": False,
            "useCache": True
        }
    }

    try:
        response = requests.post(url, json=payload, headers=headers)
        response.raise_for_status()
        result = response.json()
        print("Result is moving ahead", result)
        # Safely extract bodyBlocks
        body_blocks = (
            result.get("data", {})
            .get("getWebsiteLayoutPage", {})
            .get("pageData", {})
            .get("bodyBlocks", [])
        )
        print("\n--- Body blocks ---\n")
        print("Body blocks:", body_blocks)

        # Extract and combine all HTML text blocks
        html_blocks = [
            block["settings"]["text"]
            for block in body_blocks
            if block.get("settings", {}).get("text")
        ]
        combined_html = "\n".join(html_blocks)

        # Convert HTML to clean text
        soup = BeautifulSoup(combined_html, "html.parser")

        # Extract paragraphs into dict format
        paragraphs_dict = {
            f"p{idx + 1}": p.get_text(strip=True)
            for idx, p in enumerate(soup.select("p"))
        }
        paragraphs = [paragraphs_dict]
        print("\n--- Paragraphs ---\n")
        print("Paragraphs:", paragraphs)
        print("\n--- Date ---\n")
        print("Date", result.get("data", {}).get("getWebsiteLayoutPage", {}).get("pageData", {}).get("publishedDate"))
        print("\n--- Title ---\n")
        print("Title", result.get("data", {}).get("getWebsiteLayoutPage", {}).get("pageData", {}).get("name"))

        return {
            paragraphs: paragraphs_dict,
            "date": result.get("data", {}).get("getWebsiteLayoutPage", {}).get("pageData", {}).get("publishedDate"),
            "title": result.get("data", {}).get("getWebsiteLayoutPage", {}).get("pageData", {}).get("name"),
        }

    except requests.RequestException as e:
        print("Request failed:", e)
    except Exception as e:
        print("Error parsing article:", e)

    return None


# === Example usage ===
article_url = "https://www.energytech.com/renewables/article/21271503/cummins-integrating-hydrogen-engines-into-taylor-heavy-industrial-lift-trucks"
text = extract_article_text(article_url)

if text:
    print("\n--- Extracted Article Text ---\n")
    print(text)
else:
    print("Failed to extract article text.")
```
