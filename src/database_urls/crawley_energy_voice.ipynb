```python
import time
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
```

```python
def scrape_page(page_url):
    response = requests.get(page_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Replace 'article-container' with the actual class or ID from the HTML
    articles = soup.find_all('article', class_='post post--list')
    urls = [urljoin(page_url, article.find('a')['href']) for article in articles if article.find('a')]

    return urls

all_urls = []

for page in range(1, 27):
    page_url = f'https://www.energyvoice.com/category/renewables-energy-transition/hydrogen/page/{page}/'
    print(f'Scraping {page_url}')
    
    urls = scrape_page(page_url)
    all_urls.extend(urls)  # Add the found URLs to the all_urls list
    print(f'Found {len(urls)} URLs on page {page}')
    print('Crawley read another page ;)')
    time.sleep(1)

# Optionally, remove duplicates
all_urls = list(set(all_urls))

# Print or use the all_urls list
for url in all_urls:
    print(url)
```

```python
import pandas as pd

df = pd.DataFrame(list(set(all_urls)), columns=['URL'])
df.sort_values('URL').to_csv('URLs/energy_voice_hydrogen_urls.csv', index=False)
```
