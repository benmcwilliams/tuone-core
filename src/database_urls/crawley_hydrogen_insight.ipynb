```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

base_url = 'https://www.hydrogeninsight.com/production?page='
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

def scrape_page(page_url):
    response = requests.get(page_url, headers=headers)
    response.raise_for_status()
    soup = BeautifulSoup(response.content, 'html.parser')

    # Updated to target the correct container class based on the HTML structure
    articles = soup.find_all('div', class_='teaser-body')

    # Extracting URLs from the article links
    urls = [urljoin(base_url, article.find('a')['href']) for article in articles if article.find('a')]
    return urls

all_urls = []
for page in range(1, 31):  # Adjust the range as needed
    page_url = f'{base_url}{page}'
    print(f'Scraping {page_url}')
    
    urls = scrape_page(page_url)
    all_urls.extend(urls)
    print(f'Found {len(urls)} URLs on page {page}')
    time.sleep(1)

# Removing duplicates
all_urls = list(set(all_urls))

for url in all_urls:
    print(url)
```

```python
import pandas as pd

df = pd.DataFrame(list(set(all_urls)), columns=['URL'])
df.sort_values('URL').to_csv('URLs/hydrogen_insight_production.csv', index=False)
```
